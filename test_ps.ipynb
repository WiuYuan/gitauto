```python
import numpy as np
import matplotlib.pyplot as plt

a0 = 1  # Bohr半径单位
r = np.linspace(0, 30, 1000)

# 3s径向波函数
R3s = (2/(81*np.sqrt(3*a0**3))) * (27 - 18*r/a0 + 2*(r/a0)**2) * np.exp(-r/(3*a0))
P3s = r**2 * R3s**2

plt.plot(r, P3s)
plt.xlabel("r (a0)")
plt.ylabel("Radial Probability P(r)")
plt.title("3s Orbital Radial Probability Function")
plt.grid(True)
plt.show()
```
